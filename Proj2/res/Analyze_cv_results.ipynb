```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('cv_results.json', 'r') as f:
    results = json.load(f)
```

```python
df = pd.DataFrame(results)
df
```

```python
df['acc_train_mean'] = df.acc_train.apply(np.mean)
df['acc_train_std'] = df.acc_test.apply(np.std)
df['acc_test_mean'] = df.acc_test.apply(np.mean)
df['acc_test_std'] = df.acc_test.apply(np.std)
df
```

```python
models = df.model.unique()
fig, ax = plt.subplots(len(models), 1, sharex=True, sharey=True, figsize=(6, 12))

for i, model in enumerate(models):
    subdf = df.loc[df.model == model]
    plt.axes(ax[i])
    plt.errorbar(subdf.lr, subdf.acc_train_mean, yerr=subdf.acc_train_std, marker='o', label=f'train acc')
    plt.errorbar(subdf.lr, subdf.acc_test_mean, yerr=subdf.acc_test_std, marker='o', label=f'test acc')
    
    plt.grid(alpha=.5)
    plt.ylabel(f'{model}\naccuracy')
    if i == 1:
        plt.legend()
    
plt.ylim([0.92, 1])
plt.tight_layout()
plt.xscale('log')
plt.xlabel('lr');
```
